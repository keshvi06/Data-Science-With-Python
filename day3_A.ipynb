### 3.Bars Plot

```python
enrollment = [104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','JavaScript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid(0)
plt.bar(courses, enrollment, color='Grey')
```

```python
# For Horizontal Graph use barh & change the axis name
plt.ylabel('Courses')
plt.xlabel('Enrolled Students')
plt.grid()
plt.barh(courses, enrollment, color='Grey')
```

### 4.Pie Chart

```python
plt.figure(figsize=(10,10))
plt.pie(enrollment, labels=courses, autopct='%2.2f', shadow=True);
```

## Machine Learning

### Regression

```python
# Dataset: Salary_Data.csv, mtcars.csv, Social_network_ads.csv, Mall_Customer.csv

# Download location: http://mitu.co.in/dataset
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.shape
```

```python
df
```

```python
plt.scatter(df['YearsExperience'], df['Salary']);
```

#### -Seprate the input and output data

```python
x = df['YearsExperience']. values
y = df['Salary']
```

```python
x
```

```python
y;
```

```python
x = x.reshape(30,1)      #Reshape(rows,column)
```

```python
x
```

#### import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of linear regression

```python
regressor = LinearRegression()
```

#### train the algorithm

```python
regressor.fit(x, y)
```

#### predict on the unseen data

```python
regressor.predict([[15]])        #prediction of salary after 15 years by using predict() function
```

```python
regressor.predict([[20]])
```

```python
regressor.score(x, y)        # to find the accuracy of algorithm
```
